```python
import json
import pathlib
from pathlib import Path

#Получаем строку, содержащую путь к рабочей директории:
dir_path = pathlib.Path.cwd()
 
# Объединяем полученную строку с недостающими частями пути
path = Path(dir_path, 'Desktop','python_task', 'countries.json')

#читаем файл json
with open(path) as s:
    countries = json.load(s)

#выводим на экран тип данных (list)
print(type(countries))

#выводим на экран страны, в которых есть буква A
for countrie in countries:
    if 'a' in countrie['name'].lower():
        print(countrie['name'])
            
```
